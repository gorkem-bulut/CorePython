# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
# 1.Problemin Çözümünü Buraya Yazınız

class Vehicle():
    def __init__(self, kind):
        self.kind = kind
    
    #def showInfo(self): # Just to check
        #print(f"This is a {kind}")

class Bus(Vehicle):
    def __init__(self, kind, seating_capacity):
        super().__init__(kind)
        self.seating_capacity = seating_capacity
        
    def showInfo(self):
        print(f"The steating capacity of a {self.kind} is {self.seating_capacity}.")
    
bus1 = Bus("bus", 50)
bus1.showInfo()
````
